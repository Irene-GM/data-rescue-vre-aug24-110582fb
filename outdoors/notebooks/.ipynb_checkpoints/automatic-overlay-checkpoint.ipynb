## Automatic table overlay to ease the digitization of cells

The table overlay method used in this approach requires the user to specify how many rows the corrected journal image has. After doing this step, the `overlay.py` module will create an equally spaced table to overlay on the journal, hence sharply increasing the contrast, which is helpful for the selected digitizing methods.

### Importing libraries and setting up parameters

```python
import os
import sys
sys.path.append("/home/jovyan/work/outdoors/data_rescue_iw24/")
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps 
from pdf2image import convert_from_path
from tools.overlay import draw_table, process_images
from tools.overlay_select_points import process_images_select_points
```

### Creating the table overlay for St. Eustatius logbooks (1911)

```python
# Parameters for the table overlay methods
rows = 41
columns = 1
OVERWRITE = False
input_dir = "/home/jovyan/work/outdoors/sample-data/logbooks/corrected/st-eustatius/1911/"

# The variable will allow users to toggle between the auto-row overlay or semi-automatic row overlay methods. 
# For the semi-automatic row overlay method (process_images_select_points), the user can select the start and the end of table. 
# The function will then draw lines based on the number of rows specified by user
overlay_select_points = False

if overlay_select_points:
    # Semi-automatic row overlay
    num_lines = rows + 1
    output_dir = r"/home/jovyan/work/outdoors/sample-data/logbooks/overlayed/st-eustatius/1911-with-overlay-using-selected-points/"
    process_images_select_points(input_dir, output_dir, num_lines, OVERWRITE)
else:
    # Automatic row overlay (no user action required)
    output_dir = r"/home/jovyan/work/outdoors/sample-data/logbooks/overlayed/st-eustatius/1911-with-overlay/"
    print("Working on it!")
    process_images(input_dir, output_dir, rows, columns, OVERWRITE)

# Now we turns each of the pages in the PDF in a PNG image for visualization
print("Storing as PNG")
images = convert_from_path(os.path.join(output_dir, "1911-with-overlay.pdf"))
for i, pdf in enumerate(images):
    order = str(i + 1).zfill(2)
    pdf.save(os.path.join(output_dir, f'st-eustatius-1911_{order}.png'), 'PNG')
```

### Visualizing the effect of the automatic overlay and the semi-automatic table selection

Now we visualize what just happened

```python
# Set default figure size and DPI
plt.rcParams['figure.figsize'] = (10, 6)  # width, height in inches
plt.rcParams['figure.dpi'] = 100  # dots per inch

input_dir = "/home/jovyan/work/outdoors/sample-data/logbooks/corrected/st-eustatius/1911/"
output_dir = r"/home/jovyan/work/outdoors/sample-data/logbooks/overlayed/st-eustatius/1911-with-overlay/"
output_dir_sel_points = r"/home/jovyan/work/outdoors/sample-data/logbooks/overlayed/st-eustatius/1911-with-overlay-using-selected-points/"

file = r"2024-06-04_15-07-47_0029.png"
file_overlay = r"st-eustatius-1911_01.png"

path_in_ori = os.path.join(input_dir, file)
path_in_over = os.path.join(output_dir, file_overlay)
path_in_sel = os.path.join(output_dir_sel_points, file)

img_corr = cv2.imread(path_in_ori)
img_overlay = cv2.imread(path_in_over)
img_overlay_sel_points = cv2.imread(path_in_sel)

plt.imshow(img_corr, cmap=plt.cm.gray)
plt.title("Corrected image")
plt.show()
plt.imshow(img_overlay, cmap=plt.cm.gray)
plt.title("With overlay")
plt.show()
plt.imshow(img_overlay_sel_points, cmap=plt.cm.gray)
plt.title("With overlay using selected points")
plt.show()
```
