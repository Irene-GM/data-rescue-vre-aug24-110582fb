## Comparing historic records from St. Eustatius and St. Maarten with reanalysis data during a severe storm in Aug 1924
**Analysis:** Alessandra & Julia    |    **Visualization:** Irene    |    June 2024

This notebook illustrates the importance of rescuing historic observations from weather journals by comparing it with the reanalysis data. The cells below will prepare plots comparing time-series of `temperature`, `air pressure`, `precipitation` for St. Eustatius and St. Maarten for the month of `Aug 1924`. By the end of this month a severe storm crossed the region, leaving behing important precipitation records. [You can read more on the 1924 storm here](https://www.delpher.nl/nl/kranten/view?facets%5Btype%5D%5B%5D=artikel&facets%5Bspatial%5D%5B%5D=Nederlandse+Antillen&cql%5B%5D=%28date+_gte_+%2201-09-1922%22%29&cql%5B%5D=%28date+_lte_+%2201-10-1932%22%29&query=orkaan+AND+storm&coll=ddd&redirect=true&sortfield=date&page=1&identifier=ddd:010280908:mpeg21:a0032&resultsidentifier=ddd:010280908:mpeg21:a0032&rowid=1).

### Importing libraries

````python
import netCDF4 as nc
import xarray as xr
import glob
import pandas as pd
import os
import openpyxl

import datetime
import itertools
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

import shutup; shutup.please()
````

### Processing: Merging pre-processed data for St. Eustatius

````python
pressure_E_1924 = nc.Dataset("../sample-data/reanalysis_example/pressure_E_1924_08.nc")
temp_E_1924 = nc.Dataset("../sample-data/reanalysis_example/temp_E_1924_08.nc")
precip_E_1924 = nc.Dataset("../sample-data/reanalysis_example/precip_E_1924_08.nc")

file_pattern = '../sample-data/reanalysis_example/*E_1924_08.nc'
file_paths = glob.glob(file_pattern)
print(file_paths)

merged_1924_E = xr.open_mfdataset(file_paths)

if os.path.exists("../sample-data/reanalysis_example/merged_1924_E.nc"):
    os.remove("../sample-data/reanalysis_example/merged_1924_E.nc")
    merged_1924_E.to_netcdf("../sample-data/reanalysis_example/merged_1924_E.nc")
else:
    merged_1924_E.to_netcdf("../sample-data/reanalysis_example/merged_1924_E.nc")
````

### Processing: Merging pre-processed data for St. Maarten

````python
pressure_M_1924 = nc.Dataset("../sample-data/reanalysis_example/pressure_M_1924_08.nc")
temp_M_1924 = nc.Dataset("../sample-data/reanalysis_example/temp_M_1924_08.nc")
precip_M_1924 = nc.Dataset("../sample-data/reanalysis_example/precip_M_1924_08.nc")

file_pattern = '../sample-data/reanalysis_example/*M_1924_08.nc'
file_paths = glob.glob(file_pattern)
print(file_paths)

merged_1924_M = xr.open_mfdataset(file_paths)

if os.path.exists("../sample-data/reanalysis_example/merged_1924_M.nc"):
    os.remove("../sample-data/reanalysis_example/merged_1924_M.nc")
    merged_1924_M.to_netcdf("../sample-data/reanalysis_example/merged_1924_M.nc")
else:
    merged_1924_M.to_netcdf("../sample-data/reanalysis_example/merged_1924_M.nc")
````

### Processing: convert to dataframe and merging data for the two islands

````python
ds = xr.open_dataset('../sample-data/reanalysis_example/merged_1924_E.nc')
df_E_1924 = ds.to_dataframe().reset_index()
ds = xr.open_dataset('../sample-data/reanalysis_example/merged_1924_M.nc')
df_M_1924 = ds.to_dataframe().reset_index()
df_E_1924['Location'] = "st. Eustatius"
df_M_1924['Location'] = "st. Maarten"
merged_1924 = pd.concat([df_E_1924, df_M_1924])
````

### Clean up data

````python
merged_1924 = merged_1924.drop(['lat', 'lon', 'nbnds', 'time_bnds'], axis=1)
merged_1924 = merged_1924.drop_duplicates()
merged_1924['air'] -= 273.15
merged_1924['pres'] *= 0.00750061683
merged_1924 = merged_1924.rename(columns={'air': 'Temperature', 'pres': 'Pressure', 'apcp': 'Precipitation'})
merged_1924['day'] = merged_1924['time'].dt.day
merged_1924['hour'] = merged_1924['time'].dt.hour
merged_1924 = merged_1924.drop('time', axis=1)
merged_1924['hour'] /= 24
merged_1924['Time'] = merged_1924['day'] + merged_1924['hour']
````

````python
if os.path.exists("../sample-data/reanalysis_example/merged_1924.xlsx"):
    os.remove("../sample-data/reanalysis_example/merged_1924.xlsx")
    merged_1924.to_excel("../sample-data/reanalysis_example/merged_1924.xlsx", index=False)
else:
    merged_1924.to_excel("../sample-data/reanalysis_example/merged_1924.xlsx", index=False)
````

### Visualizing the observational measurements along with the reanalysis data (Aug 1924)

````python

# Setting up some auxiliary variables
nrows = range(0, 3)
ncols = range(0, 2)
pairs = list(itertools.product(nrows, ncols))
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(18, 15))
plt.subplots_adjust(hspace=0.5)

sd = datetime(1924, 8, 1)
ed = datetime(1924, 9, 1)
date_range_rean = list(pd.date_range(start=sd, end=ed, freq="3H"))
date_range_hist = list(pd.date_range(start=sd, end=ed, freq="D"))[:-1]

# Read historic observational data and doing some column sanity to remove "average of the day" rows,
# unify the name of the columns, and bringing pressure in St. Maarten to reasonable levels :)
df_E_obs = pd.read_excel("../sample-data/reanalysis_example/st_E.xlsx", sheet_name=0, header=0)
df_M_obs = pd.read_excel("../sample-data/reanalysis_example/st_M.xlsx", sheet_name=0, header=0)
df_M_obs["Pressure"] += 700
df_E_obs = df_E_obs[df_E_obs.Type == "Average"]
df_M_obs = df_M_obs[df_M_obs.Time == "Average"]
norm_col_names = ["Day", "Time", "Pressure", "Temperature", "Precipitation"]
df_E_obs.columns = norm_col_names
df_M_obs.columns = norm_col_names

tasks = [("St. Eustatius - Aug 1924 - Temperature", df_E_obs["Temperature"], merged_1924.loc[merged_1924["Location"]=='st. Eustatius',"Temperature"]), ("St. Maarten - Aug 1924 - Temperature", df_M_obs["Temperature"], merged_1924.loc[merged_1924["Location"]=='st. Maarten',"Temperature"]), 
        ("St. Eustatius - Aug 1924 - Precipitation", df_E_obs["Precipitation"], merged_1924.loc[merged_1924["Location"]=='st. Eustatius',"Precipitation"]), ("St. Maarten - Aug 1924 - Precipitation", df_M_obs["Precipitation"], merged_1924.loc[merged_1924["Location"]=='st. Maarten',"Precipitation"]),
        ("St. Eustatius - Aug 1924 - Pressure", df_E_obs["Pressure"], merged_1924.loc[merged_1924["Location"]=='st. Eustatius',"Pressure"]), ("St. Maarten - Aug 1924 - Pressure", df_M_obs["Pressure"], merged_1924.loc[merged_1924["Location"]=='st. Maarten',"Pressure"])]

for i in range(len(pairs)):
    r, c = pairs[i]
    title, col_hist, col_rean = tasks[i]
    ax[r,c].plot_date(date_range_hist, col_hist, fmt="-", color="red", label="Archive")
    ax[r,c].plot_date(date_range_rean, col_rean, fmt="-", color="blue", label="Reanalysis")
    ax[r,c].set_title(title)
    ax[r,c].set_xticklabels(ax[r,c].get_xticklabels(), rotation=30)
    # ax[r,c].set_xlabel("")

# Adding a single legend for the entire figure
lines_labels = [ax[r,c].get_legend_handles_labels()]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, loc="center right")
````
