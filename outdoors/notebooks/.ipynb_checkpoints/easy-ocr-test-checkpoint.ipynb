```python
import cv2
import easyocr
import matplotlib.pyplot as plt


reader = easyocr.Reader(["nl"])


def process_image(image_path, reader):
    print(f"Processing {image_path}...")
    image = cv2.imread(image_path)
    result = reader.readtext(image_path)
    process_result(result, image)

     # show the output image
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 12))
    ax.imshow(image)
    plt.show()

def process_result(result, image, probability_threshold=0.5):
    for bbox, text, prob in result:
        if prob < probability_threshold:
            continue
        draw_annotation(image, bbox, text, prob)

    # show the output image
    plt.imshow(image)
    plt.show()

def draw_annotation(image, bbox, text, prob):
    # unpack the bounding box
    (tl, tr, br, bl) = bbox
    tl = (int(tl[0]), int(tl[1]))
    tr = (int(tr[0]), int(tr[1]))
    br = (int(br[0]), int(br[1]))
    bl = (int(bl[0]), int(bl[1]))
    # cleanup the text and draw the box surrounding the text along
    # with the OCR'd text itself
    text = cleanup_text(text)
    annotation = f"{text} ({prob:.2f})"
    bg_color = (30, 200, 30)
    font = cv2.FONT_HERSHEY_PLAIN
    font_scale = 1.3
    font_thickness = 2
    text_color = (30, 30, 30)

    cv2.rectangle(image, tl, br, bg_color, 2)

    (text_width, text_height), baseline = cv2.getTextSize(
        annotation,
        font,
        font_scale,
        font_thickness,
    )
    # Define the text position
    x, y = tr[0], tr[1] + text_height + baseline
    # Draw the filled rectangle
    cv2.rectangle(
        image,
        (x, y - text_height - baseline),
        (x + text_width, y + baseline),
        bg_color,
        cv2.FILLED,
    )
    # Draw the text
    cv2.putText(image, annotation, (x, y), font, font_scale, text_color, font_thickness)


def cleanup_text(text):
    """Strip out non-ASCII text so we can draw the text on the image using OpenCV."""
    return "".join([c if ord(c) < 128 else "" for c in text]).strip()
```
